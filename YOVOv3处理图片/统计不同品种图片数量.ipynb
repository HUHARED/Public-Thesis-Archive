```python
#自己谷歌下载图片后，统计现有不同品种的图片数_
path=r"./images/cat_breeds_two/"
countRoot=0
countDirs=0
countFiles=0

for root,dirs,files in os.walk(path):
    for ele in root:
        countRoot+=1
    for ele in dirs:
        countDirs+=1
    for ele in files:
        countFiles+=1
print(countRoot)
print(countDirs)
print(countFiles)
#那么我要处理的是：分别计算65个文件夹下的文件数量
#创建DataFrame
temp=os.listdir(path)
print(len(temp))
print(type(temp))
dfBase=pd.DataFrame(temp)
dfBase=dfBase.rename(columns={0:"breed"})
dfBase["dir"]=""
dfBase["count"]=0
#填入不同文件夹的路径
seek=0
for root,dirs,files in os.walk(path):
    for ele in dirs:
        dfBase.loc[seek,"dir"]=os.path.join(path,ele)
        seek+=1
#遍历不同文件夹，计算不同文件夹下的图片文件数量
seek=0
for ele in dfBase["dir"]:
    count=0
    for root,dirs,files in os.walk(ele):
        for file in files:
            count+=1
    dfBase.loc[seek,"count"]=count
    seek+=1
#将初步统计结果保存
dfBase.to_csv("./cat_breeds_two_total_谷歌增加图片后.csv",index=False)
```

```python
#谷歌补充图片前，打印少于200张图的品种，打印还缺多少到200
dfNow=pd.read_csv('./cat_breeds_two_total_谷歌增加图片后.csv')
count =0
lowThen100=0
for ele in dfNow.iterrows():
    # temp=type(ele)
    temp=ele[1]
    if temp["count"]<200 and temp["count"]>=100:
        print(temp["breed"],"\t已有数量",temp["count"],"\t还需数量",200-temp["count"])
        count+=1
print("总共有{}种猫图片数量大于100，小于200".format(count))
for ele in dfNow.iterrows():
    temp=ele[1]
    if temp["count"]<100:
        print(temp["breed"],"图片数量小于100，\t已有数量",temp["count"])
        lowThen100+=1
print("总共有{}种猫图片数量小于100".format(lowThen100))
```
